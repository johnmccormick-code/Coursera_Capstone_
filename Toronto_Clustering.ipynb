```python
# specifying web page to be scraped 
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
# opening the url for HTML
page = urllib.request.urlopen(url)
```

```python
# extracting html data from entire web page 
wiki_soup = BeautifulSoup(page, 'lxml')
print(wiki_soup.prettify())
```

```python
# extract the table defined from the HTML starting from the "wikitable sortable"
wiki_table=wiki_soup.find("table", class_='wikitable sortable')
wiki_table
```

```python
#add data from html to new wiki_table with the three columns
Postcode = []
Borough = []
Neighborhood = []

for row in wiki_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        Postcode.append(cells[0].find(text=True))
        Borough.append(cells[1].find(text=True))
        Neighborhood.append(cells[2].find(text=True))
```

```python
#define column names while creating pandas dataframe
df=pd.DataFrame(Postcode, columns=['Postalcode'])
df['Borough']=Borough
df['Neighborhood']=Neighborhood
df
```

```python
# Data cleaning by eliminating not assigned Boroughs
df = df[df['Borough'] != 'Not assigned\n'].reset_index(drop=True)
df.head(5)
```

```python
# Grouping by Boroughs for neighborhoods
toronto_df_data= df.groupby(['Postalcode', 'Borough'], as_index = False).agg(lambda x: ", ".join(x))
toronto_df_data.head()
```

```python
# Data Cleaning by eliminting unnecessary add on characters 
toronto_df_data['Postalcode'] = toronto_df_data['Postalcode'].str.replace('\n','')
toronto_df_data['Borough'] = toronto_df_data['Borough'].str.replace('\n', '')
toronto_df_data['Neighborhood'] = toronto_df_data['Neighborhood'].str.replace('\n','')
toronto_df_data.head()
```

```python
# retrieving latitude and logitude data
df_latlog = pd.read_csv("http://cocl.us/Geospatial_data")
df_latlog.rename(columns={'Postal Code': 'Postalcode'}, inplace=True)
df_latlog.head()
```

```python
# Converting data into appropriate data types for merging purposes
toronto_df_data['Postalcode'] = toronto_df_data['Postalcode'].astype(str)
toronto_df_data['Borough'] = toronto_df_data['Borough'].astype(str)
toronto_df_data['Neighborhood'] = toronto_df_data['Neighborhood'].astype(str)
toronto_df_data.dtypes
```

```python
# Converting data into appropriate data types for merging purposes
df_latlog['Postalcode'] = df_latlog['Postalcode'].astype(str)
df_latlog['Latitude'] = df_latlog['Latitude'].astype(np.float64)
df_latlog['Longitude'] = df_latlog['Longitude'].astype(np.float64)
df_latlog.dtypes
```

```python
# Merging the two dataframes together 
toronto_newdf = pd.merge(left = toronto_df_data, right = df_latlog, left_on = 'Postalcode', right_on = 'Postalcode')
toronto_newdf.head()
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

Clean the json file and structure it into a dataframe

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
#Exploring the neighborhoods in Downtown Toronto by creating a function the repeats the processes above to the all the neighborhoods
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```
